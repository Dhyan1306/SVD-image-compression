```python
img = Image.open('LA_project.jpg').convert('L')
A = np.array(img, dtype=float)

print("Shape:", A.shape)
print("Datatype:", A.dtype)
print("Min pixel:", A.min())
print("Max pixel:", A.max())

plt.imshow(A, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()

# 8x8 heatmap
plt.imshow(A[:8, :8], cmap='hot')
plt.title("8x8 Pixel Heatmap")
plt.colorbar()
plt.show()
```

```python
U, S, Vt = np.linalg.svd(A, full_matrices=False)

print("U:", U.shape)
print("S:", S.shape)
print("Vt:", Vt.shape)

# Singular values
plt.plot(S)
plt.title("Singular Values")
plt.xlabel("Index")
plt.ylabel("Value")
plt.yscale('log')
plt.show()

# Cumulative energy
energy = np.cumsum(S**2) / np.sum(S**2)

plt.plot(energy)
plt.title("Cumulative Energy")
plt.xlabel("k")
plt.ylabel("Energy")

for t in [0.9, 0.95, 0.99]:
    k = np.argmax(energy >= t)
    plt.axhline(y=t, linestyle='--')
    plt.axvline(x=k, linestyle='--')
    print(f"{int(t*100)}% energy at k =", k)

plt.show()
```

```python
def compress_image(U, S, Vt, k):
    return np.clip(U[:, :k] @ np.diag(S[:k]) @ Vt[:k, :], 0, 255)
```

```python
k_values = [1, 5, 10, 20, 50, 100]

plt.figure(figsize=(12, 8))

for i, k in enumerate(k_values):
    Ak = compress_image(U, S, Vt, k)

    compression_ratio = (A.shape[0]*A.shape[1]) / (k*(A.shape[0]+A.shape[1]+1))
    error = np.linalg.norm(A - Ak, 'fro') / np.linalg.norm(A, 'fro')

    plt.subplot(2, 3, i+1)
    plt.imshow(Ak, cmap='gray')
    plt.title(f"k={k}\nCR={compression_ratio:.2f}\nErr={error:.3f}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
def psnr(original, compressed):
    mse = np.mean((original - compressed) ** 2)
    if mse == 0:
        return float('inf')
    return 20 * np.log10(255.0 / np.sqrt(mse))
```

```python
k_vals = [1, 2, 5, 10, 20, 30, 50, 75, 100, 150]

psnr_vals = []
cr_vals = []
err_vals = []

for k in k_vals:
    Ak = compress_image(U, S, Vt, k)

    cr = (A.shape[0]*A.shape[1]) / (k*(A.shape[0]+A.shape[1]+1))
    err = np.linalg.norm(A - Ak, 'fro') / np.linalg.norm(A, 'fro')
    p = psnr(A, Ak)

    psnr_vals.append(p)
    cr_vals.append(cr)
    err_vals.append(err)

    print(f"k={k}, CR={cr:.2f}, PSNR={p:.2f}, Error={err:.4f}")
```

```python
plt.plot(k_vals, psnr_vals, marker='o')
plt.title("PSNR vs k")
plt.xlabel("k")
plt.ylabel("PSNR")
plt.show()

plt.plot(cr_vals, psnr_vals, marker='o')
plt.title("PSNR vs Compression Ratio")
plt.xlabel("Compression Ratio")
plt.ylabel("PSNR")
plt.show()
```

```python
img_color = Image.open('LA_project.jpg')
img_arr = np.array(img_color, dtype=float)

channels = []

for i in range(3):
    Uc, Sc, Vtc = np.linalg.svd(img_arr[:, :, i], full_matrices=False)
    compressed = compress_image(Uc, Sc, Vtc, 50)
    channels.append(compressed)

compressed_img = np.stack(channels, axis=2).astype(np.uint8)

plt.imshow(compressed_img)
plt.title("Color Compressed Image")
plt.axis('off')
plt.show()
```
